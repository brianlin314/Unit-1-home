```python
import numpy as np
import pandas as pd
import warnings
```

```python
pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/SSD/p76111262/all_filtered_data.csv')
df.head()
```

```python
df.info()
```

```python
columns_to_delete = ['Flow ID', 'Src IP', 'Src Port', 'Dst IP']
df.drop(columns_to_delete, axis=1, inplace=True)
```

```python
for col in df.columns:
    # 將列轉換為浮點數
    if col != 'Label' and col != 'Timestamp':
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.info()
```

```python
numerical_fea = list(df.select_dtypes('object').columns)
numerical_fea
```

```python
columns_with_nan = df.columns[df.isnull().any()].tolist()
columns_with_nan
```

```python
df.isnull().sum().to_dict()
```

```python
df['Flow Byts/s'] = df['Flow Byts/s'].fillna(0)
df['Flow Pkts/s'] = df['Flow Pkts/s'].fillna(0)
df = df.dropna()
df.isnull().sum().to_dict()
```

```python
# 篩選出所有數值類型的列
numeric_cols = df.select_dtypes(include=[np.number])

# 檢查數值類型的欄位是否包含無限大值
columns_with_inf = numeric_cols.columns.to_series()[np.isinf(numeric_cols).any()]

# 找出包含無窮大值的行索引
rows_with_inf = numeric_cols.index[np.isinf(numeric_cols).any(axis=1)]

print("包含無窮大值的欄位:", columns_with_inf)
print("包含無窮大值的行索引:", rows_with_inf.tolist())
```

```python
# 取得float64類型可表示的最大值和最小值
max_float = np.finfo(np.float64).max
min_float = np.finfo(np.float64).min

# 僅選擇數值類型的欄位進行操作
numeric_cols = df.select_dtypes(include=[np.number])

# 遍歷所有數值列，分別取代正無窮大和負無窮大值
for col in numeric_cols:
    df[col] = df[col].replace([np.inf], max_float)
    df[col] = df[col].replace([-np.inf], min_float)

# 檢查替換是否成功
if not df.select_dtypes(include=[np.number]).applymap(np.isfinite).all().all():
    print("仍存在無窮大值或NaN。")
else:
    print("所有無窮大值已成功替換。")
```

```python
from datetime import datetime

# 將 'Timestamp' 列轉換為 datetime 類型
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%d/%m/%Y %H:%M:%S')

# 定義聚合字典
aggregations = {
     col: 'mean' for col in df.columns if col not in ['Protocol', 'Label']
}
aggregations['Protocol'] = lambda x: x.mode()[0] if not x.mode().empty else np.nan
aggregations['Label'] = lambda x: x.mode()[0] if not x.mode().empty else np.nan
aggregations['Dst Port'] = lambda x: x.mode()[0] if not x.mode().empty else np.nan

# 按 'Timestamp' 分組並套用聚合函數
df_aggregated = df.groupby('Timestamp').agg(aggregations)

# 將 'Timestamp' 列轉換為字串類型
df_aggregated['Timestamp'] = df_aggregated['Timestamp'].dt.strftime("%d/%m/%Y %H:%M:%S")

df = df_aggregated
```

```python
unique_labels = df['Label'].unique()
print(unique_labels)
```

```python
df['Label'] = df['Label'].replace("Infilteration", "Infiltration")
df['Label'] = df['Label'].replace("Bot", "Botnet")
df['Label'] = df['Label'].replace("SSH-Bruteforce", "BruteForce-SSH")
df['Label'] = df['Label'].replace("FTP-BruteForce", "BruteForce-FTP")
df['Label'] = df['Label'].replace("DoS attacks-GoldenEye", "DoS_GoldenEye")
df['Label'] = df['Label'].replace("DoS attacks-Slowloris", "DoS_Slowloris")
df['Label'] = df['Label'].replace("DoS attacks-Hulk", "DoS_Hulk")
df['Label'] = df['Label'].replace("DoS attacks-SlowHTTPTest", "DoS_SlowHTTPTest")
df['Label'] = df['Label'].replace("DDoS attacks-LOIC-HTTP", "DDoS_LOIC-HTTP")
df['Label'] = df['Label'].replace("DDOS attack-HOIC", "DDoS_HOIC")
df['Label'] = df['Label'].replace("DDOS attack-LOIC-UDP", "DDoS_LOIC-UDP")
df['Label'] = df['Label'].replace("Brute Force -XSS", "BruteForce-XSS")
df['Label'] = df['Label'].replace("SQL Injection", "SQL-Injection")
df['Label'] = df['Label'].replace("Brute Force -Web", "BruteForce-Web")

unique_labels = df['Label'].unique()
print(unique_labels)
```

```python
df.to_csv('/SSD/p76111262/CIC2018_csv/preprocess_attack.csv', index=False)

print("數據已保存到 'processed_data.csv'")
```
